```python
import os
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib.pyplot import figure
```

```python
figure(figsize=(10, 7), dpi=100)
plt.rcParams['font.family']=['Tahoma']
plt.rcParams['font.size']=10
```

```python
read_csv = pd.read_csv('../Dataset/sector_14_23_TH_.csv',encoding='utf-8', header=2)
read_csv['ภาค'][:-1].fillna(method='pad', inplace=True)
read_csv[:-1]
```

```python
def Dataset_csv():
    return read_csv['ภาค'][:-1].fillna(method='pad', inplace=True).to_csv('Dataset','.csv')

def get_new_column():
    new_column = ['ภาค']
    for indexs, rows in read_csv[:-1].iterrows():
        new_column.append(rows[1])
    new_column.append('ปี')
    return list(dict.fromkeys(new_column))

def get_value_region():
    value_region = []
    
    for indexs, rows in read_csv[:-1].iterrows():
        value_region.append(rows[0])
    return list(dict.fromkeys(value_region))

def get_region_series():
    value_region = []
    
    for indexs, rows in read_csv[:-1].iterrows():
        value_region.append(rows[0])
    df = pd.Series(list(dict.fromkeys(value_region)))
    return df

def get_value_section():
    value_section = []
    for indexs, rows in read_csv[:-1].iterrows():
        value_section.append(rows[1])
    return list(dict.fromkeys(value_section))


def get_valueof_categories_all_year(region = 'ทั่วราชอาณาจักร', category = 'ดัชนีรวม'):
    valueof_categories_all_year = []
    try:
        for indexs, rows in read_csv[:-1].iterrows():
            if region == rows[0] and category == rows[1]:
                for label, content in read_csv[:-1].items():
                    valueof_categories_all_year.append(content[indexs])
    except:
        return print('ข้อมูลไม่ถูกต้อง')
    return valueof_categories_all_year[2:]

            
def get_year():
    data_year = []
    for label, content in read_csv[:-1].items():
        data_year.append(label)
    return data_year[2:]

    
def get_data_all_category(region = 'ทั่วราชอาณาจักร'):
    
    try:
        my_data = {
        get_new_column()[1] : get_valueof_categories_all_year(region, get_new_column()[1]),
        get_new_column()[2] : get_valueof_categories_all_year(region, get_new_column()[2]),
        get_new_column()[3] : get_valueof_categories_all_year(region, get_new_column()[3]),
        get_new_column()[4] : get_valueof_categories_all_year(region, get_new_column()[4]),
        get_new_column()[5] : get_valueof_categories_all_year(region, get_new_column()[5]),
        get_new_column()[6] : get_valueof_categories_all_year(region, get_new_column()[6]),
        get_new_column()[7] : get_valueof_categories_all_year(region, get_new_column()[7]),
        get_new_column()[8] : get_valueof_categories_all_year(region, get_new_column()[8]),
        get_new_column()[9] : get_valueof_categories_all_year(region, get_new_column()[9]),
        get_new_column()[10] : get_valueof_categories_all_year(region, get_new_column()[10]),
        get_new_column()[11] : get_valueof_categories_all_year(region, get_new_column()[11]),
        get_new_column()[12] : get_valueof_categories_all_year(region, get_new_column()[12]),
        get_new_column()[13] : get_valueof_categories_all_year(region, get_new_column()[13]),
        get_new_column()[14] : get_year(),
        }
        df = pd.DataFrame(my_data, index=None)
        df.set_index('ปี',inplace=True)
    except:
        return print('ข้อมูลไม่ถูกต้อง')
    
    return df

def data_each_region(region = 'ทั่วราชอาณาจักร'):
    df = get_data_all_category(region).to_csv('data index from {}'.format(region) +'.csv')
    return df
```
