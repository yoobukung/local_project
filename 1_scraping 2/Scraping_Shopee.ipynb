```python
shopee_url = 'https://shopee.co.th/‡∏´‡πâ‡∏≠‡∏á‡∏Ñ‡∏£‡∏±‡∏ß-col.483089'

driver = webdriver.Chrome(executable_path=r'chromedriver.exe')
driver.get(shopee_url)
time.sleep(3)


def get_language_shopee():
    button_click = driver.find_element_by_css_selector('.language-selection__list-item')
    button_click.click()
    driver.implicitly_wait(3)

def get_scroll_page():
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight / 4)')
    time.sleep(3)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight / 3)')
    time.sleep(3)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight / 2)')
    time.sleep(3)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    print('scroll finish 1')

    time.sleep(5)
    driver.execute_script('window.scrollTo(document.body.scrollWidth , document.body.scrollHeight / 2)')
    time.sleep(5)
    driver.execute_script('window.scrollTo(document.body.scrollWidth, document.body.scrollHeight / 3 )')
    time.sleep(5)
    driver.execute_script('window.scrollTo(document.body.scrollWidth, document.body.scrollHeight / 4 )')
    time.sleep(5)
    print('scroll finish 2')
    
def filter_number(numb_convert):
    numb_convert.split('‡∏Ç‡∏≤‡∏¢‡πÅ‡∏•‡πâ‡∏ß')[1].split('‡∏ä‡∏¥‡πâ‡∏ô')
    if '‡∏û‡∏±‡∏ô' in numb_convert != -1:
        return float('3.7‡∏û‡∏±‡∏ô'[:-3]) * 1000
    else:
        return float(numb_convert.split('‡∏Ç‡∏≤‡∏¢‡πÅ‡∏•‡πâ‡∏ß')[1].split('‡∏ä‡∏¥‡πâ‡∏ô')[0])


def get_product_info():
    items = driver.find_elements_by_class_name('shopee-search-item-result__item')
    sell = []
    regions = []
    products = []
    
    
    for item in items:
        try:
            title_tag = item.find_element_by_tag_name('a').find_element_by_tag_name('div')
            content_tag = title_tag.find_element_by_tag_name('div').find_element_by_class_name('_1ObP5d')

            amount_sell = content_tag.find_element_by_class_name('beMRch').find_element_by_class_name('go5yPW').text
            region = content_tag.find_element_by_class_name('_2CWevj').text
            product = content_tag.find_element_by_tag_name('div').find_element_by_tag_name('div').find_element_by_tag_name('div').text
            sell.append(filter_number(amount_sell))
            regions.append(region)
            products.append(product)
        
        except:
            time.sleep(4)
            title_tag = item.find_element_by_tag_name('a').find_element_by_tag_name('div')
            content_text = title_tag.find_element_by_tag_name('div').find_element_by_class_name('_1ObP5d')
            text_content = content_text.find_element_by_class_name('beMRch').find_element_by_class_name('go5yPW').text
            region = content_tag.find_element_by_class_name('beMRch').find_element_by_class_name('_2CWevj').text
            product = content_text.find_element_by_tag_name('div').find_element_by_tag_name('div').find_element_by_tag_name('div').text
    
    return sell, regions, products

def get_table_product():
    data = {
        '‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤': get_product_info()[2],
        '‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡πÅ‡∏•‡πâ‡∏ß': get_product_info()[0],
        '‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î‡∏ó‡∏µ‡πà‡∏ï‡∏±‡πâ‡∏á': get_product_info()[1]
    }
    df = pd.DataFrame(data)
    return df
```
